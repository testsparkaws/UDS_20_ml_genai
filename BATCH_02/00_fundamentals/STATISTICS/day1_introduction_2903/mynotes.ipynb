# 📊 Statistics Notes

## Definition

**Statistics** is a domain of Mathematics/Science where we focus on **collecting, organising, analyzing & interpreting data** to make decisions.

**Examples:**
- Avg height in the class?
- Avg age in the class?
- What is the chance it will rain tomorrow?
- If I do more study, will I get better grades?


## Key Terms

| Term | Definition |
|---|---|
| **Population** | Total group (e.g., all 10 students) |
| **Sample** | Small part of population (e.g., 5 students) |
| **Parameter** | Describes the **population** (mean, median, etc.) |
| **Statistic** | Describes the **sample** |
| **Variable** | Something we measure → height, rank, marks |
| **Observation** | One data point |
| **Data** | Collection of observations (name, height, weight, age) |

## Sample Rule

> **Bigger the sample → Better the estimate**

```
1000 (marks) → sample:
    2   → 10, 10        → 10
    100 → 10,10,20,30   → ?  (50)
```

*Like taking a handful from a bag of marbles — the more you take, the better your estimate.*
