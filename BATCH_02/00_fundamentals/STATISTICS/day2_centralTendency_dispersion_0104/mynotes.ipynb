# **2.Measurement of Central Tendency**

### (1) Mean
> Goal: Find the **center** of data

**Formula:**
```
mean = Σxi / n  =  (x1 + x2 + x3) / n
```

**Example:**
```
Marks = [10, 20, 30]
mean = (10+20+30) / 3 = 60/3 = 20
```

Mean = average → single number (e.g. 50)

**Aggregate Data (Summary):**
- A → 60 → Okay
- B → 80 → Best
- C → 40 → Poor (identified by mean)

#### ⚠️ Drawback of Mean — Sensitive to Outliers

```
A → [10, 20, 40, 50]  → 120/4 = 30  ✅
B → [10, 20, 40, 90]  → 160/4 = 40
C → [10, 20, 90, 60]  → 130/4 = 32.5  → outlier effect
D → [10, 20, 90, 500] → 142   ← mean shifted by large number (500) = OUTLIER
```

### (2) Median (Middle Value)

**Example:**
```
A = [60, 50, 40, 30, 10, 50, 70, 60, 55]
sort(A) → [10, 30, 40, 40, 50, 50, 55, 60, 60]   len() = 9
                              ↑
                           median
```

**Rules:**
- **Odd** count → take the middle number directly → median
- **Even** count → take 2 centre values → avg → median

**Even example:**
```
sort(A) → [10,30,40,40,50 | 50,55,60,60,80]
value 1 & 2 → (50+50)/2 → 50 → median
```

**Outlier-resistant example:**
```
sort(A) → [10,30,40,40,50,50,55,60,60,8000]
                          ↑
                    median = 50  (8000 doesn't affect it!)
```

> ✅ If I want a **true center** of my data → use **mean / median**
> ✅ If I want an **aggregation that shows behavior** of my data → use **mean / median**


### (3) Mode → Categorical

**Example — Sports results (A, B, C, D):**
```
1→A, 2→B, 3→A, 4→B, 5→A
6→B, 7→C, 8→C, 9→C, 10→C
```
- Mean → ❌ (can't average categories)
- Median → ❌
- **Mode → ✅** "Who won the most games?"

**Frequency of categories:**
- A : 3
- B : 3
- C : **4 ← MODE** (most frequent)

**Marks → 1-10, 60 students:**
```
A  B  C  D  F
1  2  ?  7  8
         ↓
         8  ← MODE
```

> **Mode = Most Frequent Value**
> - Numerical data → numerical mode
> - Categorical data → categorical mode

**Q. [10, 20, 30, 1000]**
- Centre → **median** (recommended)
- Mean → sensitive to outlier (1000)

# **3.Measurement of Dispersion (Spread of Data)**

> **Spread = 2 hands = Standard Deviation**

**Intuition:** What is the common spread between multiple people's heights?

| Person | A | B | C | D | E |
|--------|-----|-----|-----|-----|-----|
| Height (cm) | 170 | 180 | 100 | 140 | 190 |

→ Common spread = **32.6**
→ On average, the difference between people's height is **32.6**

```
     ←32.6     mean     +32.6→
        [person]  [person]  [person]
```

> **SD = How far each point is from the mean**

### Variance & Standard Deviation — Formula

**Series = [10, 20, 30]**
```
mean = (10+20+30)/3 = 20
```

**Variance (σ²):**
```
σ² = Σ(xi − μ)² / n      ← lower case sigma squared

= avg of (difference of each point from mean)²

(10−20)² = (−10)² = 100
(20−20)² = (0)²   = 0
(30−20)² = (10)²  = 100

σ² = (100 + 0 + 100) / 3 = 66.67 unit²
```

**Standard Deviation (σ):**
```
σ = √variance = √66.67 ≈ 8.16

σ² → σ  (goal)
σ² → √σ² → σ
```

**Interpretation for [10, 20, 30], SD = 8.16:**
> "On average, numbers deviate from the mean by **8.16**"

```
Average spread:
[  10        20        30  ]
       −8.16    +8.16
      11.84     20      28.16
```

### Why SD over Variance?

| | Variance | SD |
|---|---|---|
| Unit | x inch² ❌ | x inch ✅ |
| Interpretable | No | Yes |
| Same unit as data | No | **Yes** |

* 

> SD is in the **same unit as the data** → preferred for interpretation
