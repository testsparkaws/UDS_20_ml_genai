```python
import numpy as np
from scipy import stats
```

```python
# CI
```

```python
sample_mean = 31
pop_sd = 4
n = 36
confidence = 0.95
```

```python
# SE
se = pop_sd / np.sqrt(n)
```

```python
# Calculate the Interval
lower, upper =stats.norm.interval(confidence, loc=sample_mean, scale=se)
```

```python
print(f"95% CI --- lower-bound : {lower:.4f}, upper-bound : {upper:.4f}")
```

# test

```python
bulb_data = [980, 960, 1020, 940, 950]
claimed_mean = 1000

t_stat, p_val = stats.ttest_1samp(bulb_data, claimed_mean)

print(f"T-statistic: {t_stat:.2f}")
print(f"P-value: {p_val:.4f}")
```
