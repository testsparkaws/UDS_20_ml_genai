# 🟢 **2. Quartiles & IQR (Very Important)**

---

## 📌 Quartiles

Data divided into **4 equal parts**:

| Quartile | Meaning                  |
| -------- | ------------------------ |
| Q1       | 25th percentile          |
| Q2       | 50th percentile (Median) |
| Q3       | 75th percentile          |

---

## 📌 IQR (Interquartile Range)

```text
IQR = Q3 - Q1
```

👉 Represents **middle 50% of data**

---

## 📊 Diagram Explanation (Page 1)

The box diagram shows:

* Left → Q1
* Center → Q2 (median)
* Right → Q3
* Width → IQR

👉 This is basically a **box plot**

---

# 🟢 **3. Outliers – Fence Method**

---

## 📌 Formula

Lower Fence:

```text
Q1 - 1.5 × IQR
```

Upper Fence:

```text
Q3 + 1.5 × IQR
```

👉 Any value outside this range = **Outlier**

---

## 📊 Diagram (Page 1)

* Middle box → normal data
* Points outside → outliers

---

# 🟢 **5. Numerical Example (Page 2)**

Dataset:

```text
4, 8, 12, 15, 18, 21, 24, 27, 30, 33, 38, 42, 47, 51, 94
```

---

## Step 1: Q1 (25th percentile)

Position formula:

```text
L = 0.25 × (n-1) + 1
```

Result:
👉 Q1 ≈ **16.5**

---

## Step 2: Q3 (75th percentile)

```text
L = 0.75 × (n-1) + 1
```

Result:
👉 Q3 ≈ **40**

---

## Step 3: IQR

```text
IQR = 40 - 16.5 = 23.5
```

---

## Step 4: Fences

Lower:

```text
16.5 - (1.5 × 23.5) = -18.75
```

Upper:

```text
40 + (1.5 × 23.5) = 75.25
```

---

## 📊 Diagram (Page 3)

* Values outside 75.25 → Outliers (94)

# 🟢 **9. Skewness (Page 5–6)**

---

## 📌 Definition

👉 Shows **data asymmetry**

---

## 📊 Case 1: Balanced (Normal)

* Mean ≈ Median
* Symmetrical curve

Example:

```text
48, 50, 60, 70, 80
```

---

## 📊 Case 2: Right Skew (Positive Skew)

* Tail on right side
* Mean > Median

Example:

```text
10, 20, 30, 40, 90
Mean = 38, Median = 30
```

---

## 📊 Case 3: Left Skew (Negative Skew)

* Tail on left side
* Mean < Median

Example:

```text
10, 60, 70, 80, 90
Mean = 62, Median = 70
```

---

## 📊 Diagram Explanation (Page 6)

| Shape      | Meaning       |
| ---------- | ------------- |
| Balanced   | Mean = Median |
| Right skew | Mean > Median |
| Left skew  | Mean < Median |

---

# 🟢 **10. Skewness Formula (Page 7)**

```text
Skewness = Σ (xi - μ)³ / (n × σ³)
```

---

## 📌 Terms:

* xi → data point
* μ → mean
* n → total values
* σ → standard deviation

---

## 📌 Interpretation:

* > 0 → Right skew
* < 0 → Left skew
* ≈ 0 → Normal

---

# 🟢 **11. Kurtosis (Page 7)**

---

## 📌 Meaning

👉 Measures **extreme values (tails)**

---

## 📊 Types:

| Type   | Meaning             |
| ------ | ------------------- |
| High   | Many extreme values |
| Low    | Flat distribution   |
| Normal | Balanced            |

---

## 📌 Formula

```text
Kurtosis = Σ (xi - μ)⁴ / (n × σ⁴)
```

---

# 🟢 **12. Key Insight (Page 8)**

👉 Squaring values removes direction (+/-)

---

## 📊 Interpretation:

| Value | Meaning          |
| ----- | ---------------- |
| ≈ 0   | Normal           |
| > 0   | Extreme outliers |
| < 0   | Few outliers     |
