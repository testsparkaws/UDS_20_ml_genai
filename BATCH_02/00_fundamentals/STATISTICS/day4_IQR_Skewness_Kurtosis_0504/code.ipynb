# **Outlier Detection**

```python
def detect_outliers(data, k=1.5):

    # sort data
    data.sort()

    arr = np.array(data, dtype=float)

    # Q1 & Q3
    Q1 = np.percentile(arr, 25, method='linear')
    Q3 = np.percentile(arr, 75, method='linear')

    # IQR
    IQR = Q3-Q1

    # Lower Fence & Upper Fence
    lower = Q1 - k * IQR
    upper = Q3 + k * IQR

    mask = (arr >= lower) & (arr <= upper) # to remove outliers
    outliers_mask = ~mask # to extract outlier

    return {
        "Q1" : float(Q1),
        "Q3" : float(Q3),
        "IQR" : float(IQR),
        "lower_fence" : float(lower),
        "upper_fence" : float(upper),
        "outliers" : arr[outliers_mask].tolist(),
        'clean_data' : arr[mask].tolist()
    }

# lower = 20
# upper = 90
# (value > 20) & (value < 90):
# False, False, True, ..... False, False --> mask
```

```python
student_score = [10, 12, 45, 34, 20, 33, 35, 40, 55, 44, 48, 53, 90, 98]
print(detect_outliers(student_score))
```

# **kurtosis**

```python
from scipy.stats import kurtosis
```

```python
# kurtosis
data = [10, 20, 30, 40, 1000]

mean = np.mean(data)
std = np.std(data)

skew = np.mean((data - mean)**3) / (std**3)

# Pearson's Kurtosis (Normal distribution = 3.0)
kur = np.mean((data - mean)**4) / (std**4)

# # Excess Kurtosis (Normal distribution = 0.0)
excess_kurt = (np.mean((data - np.mean(data))**4) / (np.std(data)**4)) - 3

print("Skewness : ", skew)

print("Kurtosis : " , kur)
print("Kurtosis : " , excess_kurt)

print(kurtosis(data)) 
# 3.24 extreme value exist
```

# **Extra**

```python
import pandas as pd
import numpy as np

def get_outliers_summary(df):
    # 1. Select only numeric columns (ignore strings/objects)
    numeric_df = df.select_dtypes(include=[np.number])
    summary = []
    
    for col in numeric_df.columns:
        # 2. Calculate IQR components
        Q1 = numeric_df[col].quantile(0.25)
        Q3 = numeric_df[col].quantile(0.75)
        IQR = Q3 - Q1
        
        # 3. Calculate Fences
        lower = Q1 - 1.5 * IQR
        upper = Q3 + 1.5 * IQR
        
        # 4. Count outliers
        count = ((numeric_df[col] < lower) | (numeric_df[col] > upper)).sum()
        count1  =len(df.query(f" `{col}`< @lower  or `{col}`> @upper"))
        
        
        summary.append({
            'Column': col,
            'Lower Fence': round(lower, 2),
            'Upper Fence': round(upper, 2),
            'Outlier Count': count,
            'Outlier Count using df query':count1
        })
    
    return pd.DataFrame(summary)
```
