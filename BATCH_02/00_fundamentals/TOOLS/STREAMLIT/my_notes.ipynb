# **1. Page Setup & Text Display**


### ✅ `st.set_page_config()`

👉 Used to configure page settings (title, icon, layout)

```python

st.set_page_config(page_title="My App",page_icon="🚀",layout="wide")
```

✔️ Use case: Dashboard apps (wide layout)


### ✅ `st.title()`, `st.header()`, `st.subheader()`

👉 Used for headings

```python

st.header("Section Header")
st.subheader("Sub Section")
```

✔️ Use case: Structure your UI like a webpage

👉 Supports **Markdown + HTML**


👉 Most flexible function (can print text, dataframe, charts)

✔️ Use case: Quick debugging / display anything

# **10. `st.session_state` (VERY IMPORTANT)**

👉 Used to **store data across interactions**


### ❌ Problem without session_state

Streamlit reruns script on every click → data resets


### ✅ Solution: `session_state`

```python
if "count" not in st.session_state:
    st.session_state.count = 0

if st.button("Increment"):
    st.session_state.count += 1

st.write("Count:", st.session_state.count)
```

### ✅ Store form data

```python
name = st.text_input("Enter name")

if st.button("Save"):
    st.session_state.name = name

st.write("Saved Name:", st.session_state.get("name"))
```

### 🔥 Key Points

* Acts like a dictionary
* Persists across reruns
* Essential for real apps
