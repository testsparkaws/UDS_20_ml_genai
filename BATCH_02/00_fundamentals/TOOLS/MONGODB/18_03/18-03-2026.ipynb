# SELECT

```python
# relational : selct * from dummy0;
# * --> {}
for i in collection.find({}):
    print(i)
```

```python
# find all the category which belongs to smartphone
query = {
    "category" : "smartphone"
}
for i in collection.find(query):
    print(i)
```

```python
# find all the rows which has category key in it
query = {
    "category" : { "$exists" :True}
}
for i in collection.find(query):
    print(i)
```

```python
# get all the smartphone category where price is greater than 0

query = {
    "category" : "smartphone",
    "price" : {"$gt" : 0}
    # "price" >  0
}
for i in collection.find(query):
    print(i)
```

# UPDATE

```python
query_filter = {
    "name" : {"$regex" : "17", "$options" : "i"}
}

update_operation = {
    "$set" : {
        "category" : "smartphone_latest"
    }
}

result = collection.update_many(query_filter, update_operation)

print(f"Matched {result.matched_count} documents and modified {result.modified_count} documents.")
```
