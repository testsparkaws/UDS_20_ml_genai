```python
import os
import psycopg2
from dotenv import load_dotenv
```

```python
load_dotenv()
```

```python
# Fetch variables
USER = os.getenv("user")
PASSWORD = os.getenv("password")
HOST = os.getenv("host")
PORT = os.getenv("port")
DBNAME = os.getenv("dbname")
```

```python
# Connect to the database
try:
    connection = psycopg2.connect(
        user=USER,
        password=PASSWORD,
        host=HOST,
        port=PORT,
        dbname=DBNAME
    )
except Exception as e:
    print(f"Issue with DB connection : {e}")
```

```python
cursor = connection.cursor()
```

```python
cursor.execute("SELECT * FROM student;")
result = cursor.fetchall()
print(result)
```

```python
for r in result:
    print(r)
```

```python
student_val = ("Pragati", 20, "Delhi", 9)
cursor.execute('''
    INSERT INTO student (name, age, city, class)
    VALUES (%s,%s,%s,%s)
    ''', student_val)
```

```python
connection.commit()
```
