```python
import sqlite3

# https://sqliteonline.com/
```

```python
# step.1 
conn = sqlite3.connect('student_1.db')
```

```python
#step.2
cursor = conn.cursor()
```

```python
cursor.execute('PRAGMA foreign_keys = ON;')
```

```python
cursor.execute('''
    CREATE TABLE ultimate_data_science (
        student_id INT PRIMARY KEY NOT NULL,
        name TEXT NOT NULL,
        batch_id INT NOT NULL
    )
''')
```

```python
# insert values using tuples

student_vals = [(1, "Agi Ann Maria", 2), (2, "Tanmay", 2), (3, "satyajit roy", 2)]
```

```python
for student_val in student_vals:
    print(student_val)
    cursor.execute('''
    INSERT INTO ultimate_data_science (student_id, name, batch_id)
    VALUES (?,?,?)
    ''', student_val)
```

```python
# Save --> push the data from memory to the hard-drive
# saves the state of database
conn.commit()
```

```python
conn.close()
```
