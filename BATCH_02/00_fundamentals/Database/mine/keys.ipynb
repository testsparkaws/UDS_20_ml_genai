These are **important database key concepts** used in **table design and normalization**. I will explain them step-by-step with a simple example.

Assume we have a table **`employees`**:

| emp_id | email                                   | phone     | name  | dept |
| ------ | --------------------------------------- | --------- | ----- | ---- |
| 101    | [rahul@mail.com](mailto:rahul@mail.com) | 999111222 | Rahul | IT   |
| 102    | [amit@mail.com](mailto:amit@mail.com)   | 999111333 | Amit  | HR   |
| 103    | [neha@mail.com](mailto:neha@mail.com)   | 999111444 | Neha  | IT   |



# 1️⃣ Super Key

### Definition

A **Super Key** is **any column or combination of columns that uniquely identifies a row** in a table.

A super key **may contain extra columns**.

### Example

In the `employees` table, these are **Super Keys**:

* `emp_id`
* `email`
* `phone`
* `(emp_id, name)`
* `(emp_id, email)`
* `(email, phone)`

Why?

Because **each of these combinations uniquely identifies a row**.

Example:

```
emp_id = 101
```

returns exactly **one employee**.

Even this works:

```
(emp_id, name) = (101, Rahul)
```

But **name is unnecessary**, because `emp_id` alone is enough.

👉 Therefore `(emp_id, name)` is **still a super key but not minimal**.


# 2️⃣ Candidate Key

### Definition

A **Candidate Key** is a **minimal super key**.

Meaning:

* It **uniquely identifies rows**
* **No extra columns are included**

### Example

From the previous super keys:

Possible candidate keys:

* `emp_id`
* `email`
* `phone`

Why?

Because each one **uniquely identifies an employee**.

But these are **not candidate keys**:

```
(emp_id, name)
(email, phone)
```

Because they contain **extra columns**.

So:

| Key            | Type           |
| -------------- | -------------- |
| emp_id         | Candidate Key  |
| email          | Candidate Key  |
| phone          | Candidate Key  |
| (emp_id, name) | Super Key only |


# 3️⃣ Primary Key

### Definition

A **Primary Key** is the **candidate key chosen by the database designer to uniquely identify rows**.

Important properties:

* **Unique**
* **Not NULL**
* **Only ONE primary key per table**

### Example

From the candidate keys:

```
emp_id
email
phone
```

We choose **emp_id** as the **Primary Key**.

Example table:

```sql
CREATE TABLE employees (
    emp_id INT PRIMARY KEY,
    email VARCHAR(100),
    phone VARCHAR(15),
    name VARCHAR(50),
    dept VARCHAR(50)
);
```

Now:

```
emp_id = 101
```

uniquely identifies the employee.


# 4️⃣ Surrogate Key

### Definition

A **Surrogate Key** is an **artificial key generated by the system**, not derived from business data.

Examples:

* Auto Increment ID
* UUID
* Sequence number

It has **no business meaning**.

---

### Example

Instead of using `email` or `phone`, we generate an **artificial ID**:

| emp_key | email                                   | phone     | name  |
| ------- | --------------------------------------- | --------- | ----- |
| 1       | [rahul@mail.com](mailto:rahul@mail.com) | 999111222 | Rahul |
| 2       | [amit@mail.com](mailto:amit@mail.com)   | 999111333 | Amit  |
| 3       | [neha@mail.com](mailto:neha@mail.com)   | 999111444 | Neha  |

Here:

```
emp_key
```

is a **Surrogate Key**.

SQL example:

```sql
CREATE TABLE employees (
    emp_key INT IDENTITY(1,1) PRIMARY KEY,
    email VARCHAR(100),
    phone VARCHAR(15),
    name VARCHAR(50)
);
```

Let’s continue with the same **employees example** to understand **Alternate Key** and **Foreign Key** clearly.

# 1️⃣ Alternate Key

### Definition

An **Alternate Key** is a **candidate key that is NOT chosen as the primary key**.

Recall:

* **Candidate Keys** → All possible minimal keys that uniquely identify rows.
* **Primary Key** → One of those candidate keys selected as the main identifier.
* **Alternate Keys** → The remaining candidate keys.

---

### Example

Table: `employees`

| emp_id | email                                   | phone     | name  |
| ------ | --------------------------------------- | --------- | ----- |
| 101    | [rahul@mail.com](mailto:rahul@mail.com) | 999111222 | Rahul |
| 102    | [amit@mail.com](mailto:amit@mail.com)   | 999111333 | Amit  |
| 103    | [neha@mail.com](mailto:neha@mail.com)   | 999111444 | Neha  |

Possible **Candidate Keys**

```
emp_id
email
phone
```

Suppose we choose:

```
emp_id → PRIMARY KEY
```

Then:

```
email
phone
```

become **Alternate Keys**.

---

### SQL Example

```sql
CREATE TABLE employees (
    emp_id INT PRIMARY KEY,
    email VARCHAR(100) UNIQUE,
    phone VARCHAR(15) UNIQUE,
    name VARCHAR(50)
);
```

Here:

| Column | Key Type      |
| ------ | ------------- |
| emp_id | Primary Key   |
| email  | Alternate Key |
| phone  | Alternate Key |

Because **UNIQUE constraints enforce alternate keys**.

---

# 2️⃣ Foreign Key

### Definition

A **Foreign Key** is a column that **references the Primary Key of another table**.

Purpose:

* Create **relationship between tables**
* Maintain **referential integrity**

---

### Example

Table 1: `departments`

| dept_id | dept_name |
| ------- | --------- |
| 10      | IT        |
| 20      | HR        |
| 30      | Finance   |

Primary Key

```
dept_id
```

---

Table 2: `employees`

| emp_id | name  | dept_id |
| ------ | ----- | ------- |
| 101    | Rahul | 10      |
| 102    | Amit  | 20      |
| 103    | Neha  | 10      |

Here:

```
dept_id
```

in **employees** is a **Foreign Key** referencing **departments.dept_id**

---

### SQL Example

```sql
CREATE TABLE departments (
    dept_id INT PRIMARY KEY,
    dept_name VARCHAR(50)
);
```

```sql
CREATE TABLE employees (
    emp_id INT PRIMARY KEY,
    name VARCHAR(50),
    dept_id INT,
    FOREIGN KEY (dept_id)
        REFERENCES departments(dept_id)
);
```

---

### What Foreign Key Ensures

You **cannot insert invalid references**.

❌ Invalid insert

```sql
INSERT INTO employees VALUES (104, 'Ravi', 50);
```

Because **dept_id = 50 does not exist**.

---

### Relationship Diagram

```
departments
-------------
dept_id (PK)
dept_name

      |
      |  (Foreign Key)
      v

employees
-------------
emp_id (PK)
name
dept_id (FK)
```

# 🔑 Summary Table

| Key Type      | Meaning                                         | Example           |
| ------------- | ----------------------------------------------- | ----------------- |
| Primary Key   | Main unique identifier                          | emp_id            |
| Candidate Key | Possible unique identifiers                     | emp_id, email     |
| Alternate Key | Candidate key not chosen as PK                  | email             |
| Super Key     | Any column combination uniquely identifying row | (emp_id,name)     |
| Surrogate Key | Artificial key                                  | auto_increment id |
| Foreign Key   | Reference to another table's PK                 | dept_id           |

fact_sales

| sales_key | product_id | customer_id | amount |
| --------- | ---------- | ----------- | ------ |

* `sales_key` → **Surrogate Key**

Data warehouses prefer **surrogate keys** because:

* faster joins
* stable keys
