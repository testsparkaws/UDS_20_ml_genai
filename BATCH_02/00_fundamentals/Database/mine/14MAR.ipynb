```python
cursor.execute('''
    CREATE TABLE ultimate_data_science(
      student_id INT PRIMARY KEY NOT NULL,
      name TEXT NOT NULL,
      batch_id INT NOT NULL  
    )
''')
```

```python
# Insert valies using tuples 
## 
#student_vals = (1,'Agi Ann Maria', 2)
student_vals = [
        (1,'Agi Ann Maria', 2),
        (2,'Tanmay', 2),
       (3,'Satayjit roy', 2),
]
```

```python
### INSERT 
for student_val in student_vals:
    #print(student_val)
    cursor.execute('''
        insert into ultimate_data_science (student_id,name, batch_id) 
        VALUES (?,?,?)
    ''', student_val)
```

```python
###  Save --> push the data from memory to the hard-drive
conn.commit()
```
