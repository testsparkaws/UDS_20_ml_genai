## Problem 2 — Mutable default bug
Implement a `Memo` class that stores notes per instance; ensure using `default_factory` or instance init to avoid shared lists.

```python
# Starter
class Memo:
    def __init__(self, notes=None):
        # TODO: ensure each instance has its own list
        pass

m1 = Memo()
m2 = Memo()
m1.notes.append('note1')
print('m1', m1.notes)
print('m2', m2.notes)  # should be empty list
```

```python
# Solution
class Memo:
    def __init__(self, notes=None):
        if notes is None:
            notes = []
        self.notes = notes

m1 = Memo()
m2 = Memo()
m1.notes.append('note1')
print('m1', m1.notes)
print('m2', m2.notes)
```

## Problem 3 — `@classmethod` alternate constructor
Add a `from_string` classmethod to `Point` that parses `
` into a `Point` instance.

```python
# Starter & Solution
class Point:
    def __init__(self, x, y):
        self.x = x
        self.y = y

    @classmethod
    def from_string(cls, s):
        x_str, y_str = s.split(',')
        return cls(float(x_str), float(y_str))

p = Point.from_string('3.5,4')
print(p.x, p.y)
```

## Problem 4 — Dunders: make a `Range2` like `range` with `__iter__` and `__len__`
Implement a simple `Range2(start, stop)` with iteration and length.

```python
# Solution
class Range2:
    def __init__(self, start, stop):
        self.start = start
        self.stop = stop

    def __iter__(self):
        n = self.start
        while n < self.stop:
            yield n
            n += 1

    def __len__(self):
        return max(0, self.stop - self.start)

r = Range2(2, 6)
print(list(r), len(r))
```

## Problem 5 — Dataclass with mutable field
Create a `@dataclass` `Team` with a `members: list[str]` that uses `default_factory` to ensure separate lists.

```python
# Solution
from dataclasses import dataclass, field

@dataclass
class Team:
    name: str
    members: list[str] = field(default_factory=list)

t1 = Team('Red')
t2 = Team('Blue')
t1.members.append('Alice')
print(t1, t2)
```
