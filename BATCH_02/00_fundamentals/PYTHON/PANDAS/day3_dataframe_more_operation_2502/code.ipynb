# 2. reindex() — Conform to a new index/column structure

```python
sales_jan = pd.DataFrame({
    'Sales': [250, 400, 300]
}, index=['Store_A', 'Store_B', 'Store_C'])

sales_jan
```

```python
sales_feb = pd.DataFrame({
    'Sales': [260, 310]
}, index=['Store_A', 'Store_C'])

sales_feb
```

❌ Problem: Indexes are mismatched!

You cannot safely do operations like subtraction or direct comparison unless indexes are aligned.

✅ 🛠 Solution: Use reindex() to align!

```python
common_index = sales_jan.index

print(common_index)
```

```python
sales_feb_aligned = sales_feb.reindex(common_index)

sales_feb_aligned
```

```python
sales_diff = sales_feb_aligned['Sales'] - sales_jan['Sales']

sales_diff
```

```python
sales_feb_aligned = sales_feb.reindex(common_index, fill_value=0)

sales_feb_aligned
```

```python
sales_diff = sales_feb_aligned['Sales'] - sales_jan['Sales']

sales_diff
```

# **4.Iteration in Pandas**

```python
df = pd.DataFrame({
    'Name': ['Alice', 'Bob', 'Charlie'],
    'Age': [25, 30, 35],
    'City': ['Delhi', 'Mumbai', 'Chennai']
})
```

```python
df
```

```python
for index, row in df.iterrows():
    print(f"{index}: {row['Name']} lives in {row['City']} and is {row['Age']} years old.")
```

```python
for row in df.itertuples():
    print(row)
    print(row.Index)
    print(f"{row.Name} lives in {row.City}, and is {row.Age} years old.", end='\n\n')
```

```python
# Iterating Over Columns

for col in df.columns:
    print(f"Column: {col}")
    print(df[col].values, end='\n\n')
```

# **5.Conditional Example: Add Column with Custom Logic**

```python
categories = []

# Loops

for idx, row in df.iterrows():
    print(row)
    print(row['Age'])

    if row['Age'] < 40 :
        categories.append("young")
    else :
        categories.append("senior")

df['Category'] = categories

df.head()
```

# **6.Better Way: Vectorized / Apply**

```python
df['Category_vectorized_1'] = df['Age'].apply(lambda age : 'young' if age < 40 else 'senior')

df['Category_vectorized_1'].head()
```

```python
categorize_age_lambda = lambda age : 'young' if age < 40 else 'senior'

df['Category_vectorized_2'] = df['Age'].apply(categorize_age_lambda)

df['Category_vectorized_2'].head()
```

```python
def categorize_age_function(age):
    # if age < 40:
    #     return 'young'
    # else :
    #     return 'senior'
    return 'young' if age < 40 else 'senior'

df['Category_vectorized_3'] = df['Age'].apply(categorize_age_function)

df['Category_vectorized_3'].head()
```

# **7.Sorting in Pandas**

## **1. sort_values() → Sort by data values**

```python
# Example of inplace

class Cal:
    def __init__(self, a):
        self.a = a
    def add(self, b):
        c = self.a + b
        return c
    def inplace_add(self, b):
        self.a = self.a + b
        return self.a
```

```python
c_cal = Cal(10)
```

```python
c_cal.a
```

```python
c_cal.add(10)
```

```python
c_cal.a = c_cal.add(10) # inplace False
```

```python
c_cal.a
```

```python
c_cal.inplace_add(10) # inplace True
```

```python
c_cal.a 
```
