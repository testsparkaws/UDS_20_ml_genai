### 🔷 `array.shape` in NumPy

`shape` tells you the **dimensions (size) of an array** — i.e., how many rows, columns, etc.


## ✅ Basic Idea

> `array.shape` → returns a **tuple** representing the size along each dimension


## 🔢 Example 1: 1D Array

```python
import numpy as np

a = np.array([1, 2, 3, 4])
print(a.shape)
```

✅ Output:

```
(4,)
```

👉 Meaning:

* 4 elements in one dimension


## 🔢 Example 2: 2D Array (Matrix)

```python
a = np.array([
    [1,2,3],
    [4,5,6]
])

print(a.shape)
```

✅ Output:

```
(2, 3)
```

👉 Meaning:

* 2 rows
* 3 columns

## 🔢 Example 3: 3D Array

```python
a = np.array([
    [[1,2],[3,4]],
    [[5,6],[7,8]]
])

print(a.shape)
```

✅ Output:

```
(2, 2, 2)
```

👉 Meaning:

* 2 blocks
* each block → 2 rows, 2 columns

## 🔥 How to Read Shape

| Shape     | Meaning                    |
| --------- | -------------------------- |
| `(4,)`    | 1D array with 4 elements   |
| `(2,3)`   | 2 rows, 3 columns          |
| `(2,2,2)` | 3D (blocks, rows, columns) |

## 🔄 Using `shape` to Reshape

```python
a = np.array([1,2,3,4,5,6])

b = a.reshape((2,3))
print(b)
```

## ⚠️ Important Rule

👉 Total elements must match:

```
2 × 3 = 6 ✅
```

## 🎯 Interview Tips

* `shape` is a **tuple**
* `len(shape)` → gives number of dimensions
* Use with:

  * `reshape()`
  * `ndim`
  * `size`


## 🔥 Related Attributes

```python
a.shape   # dimensions
a.ndim    # number of dimensions
a.size    # total elements
```

## 🧠 One-line Summary

> `shape` = **structure of array (rows, columns, dimensions)**
