```python
import time
```

```python
def fetch(site, secs):
    time.sleep(secs)
    return f"got {site}"
```

```python
t0 = time.time()
fetch("site-A", 1)           # wait 1s…
fetch("site-B", 2)           # then wait 2s…
fetch("site-C", 1)           # then wait 1s…
print(time.time() - t0)     # ≈ 4.0 seconds total
```

```python
import asyncio
```

```python
async def fetch(site, secs):
    await asyncio.sleep(secs) # yields — other tasks can run!
    return f"got {site}"

async def main():
    t0 = asyncio.get_event_loop().time()
    results = await asyncio.gather(
        fetch("site-A", 1),    # all 3 start simultaneously
        fetch("site-B", 2),
        fetch("site-C", 1),
    )
    print(asyncio.get_event_loop().time() - t0)  # ≈ 2 seconds!

asyncio.run(main())
```

```python
#How the event loop works
The event loop is a scheduler — it keeps a queue of coroutines and runs them one at a time. When a coroutine hits await, it pauses and the loop picks up the next ready one. Press "step" to walk through it.
```
