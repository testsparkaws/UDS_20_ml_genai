## 🔷 OOP — Polymorphism

Three separate classes, each with their own start method:

```python
class Tv:
    def start_tv(self):
        # logic to start tv

class Fan:
    def start(self):
        # logic to start fan

class Microwave:
    def start_microwave(self):
        # logic to start mv
```

### Usage:
```python
tv = Tv()
tv.start_tv()

fan = Fan()
fan.start()     # → start

mv = Microwave()
mv.start_microwave()
```

> **Problem shown:** Each class has a *different method name* for the same action (start). Polymorphism solves this by giving them a common interface.

## 📦 Modules

> **Definition:** A Python file (`.py`) containing code that can be imported.

### Example — `g.py` structure:
```
Electronic appliances  → 5 classes
Mathematical functions (bmi, c→f) → 10 functions
```

Split into separate modules:
```
math.py   → module → math functions or classes
utils.py  → open txt, append txt
electronic.py → 5 classes
```

### Built-in modules:
```
os    → implemented OS level instructions  } module
time  → time(), perf_()
```

#### Module Continued
```
1. add : x   y
   add_(...) = a() : val  y
```

> **Modules** → `.py` → code
> *"To separate complex & reusable code from the main Python file."*

### Example files:

**`calculator.py`** (used in `g.py`):
```python
import time

def hello_world():
    print("hello world")

class Calculator:
    def __init__(self):
        self.__history = []

    def add(self, *args):
        total = 0
        for i in args:
            total += i

        t = time.time()
        self.__history.append({f"Add_{t}" : {str(args) : total}})
        return total

    def get_history(self):
        print(self.__history)
```

**`temp.py`** (imports from `calculator.py`):
```python
import os  # link → os.py

os.getcwd()

c = calculator.Calculator()
c.add(1, 2)
c.get_history()
```
