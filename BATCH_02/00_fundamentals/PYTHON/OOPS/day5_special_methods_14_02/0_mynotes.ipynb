Great question! Let me show you exactly what's happening in memory.## What's happening — name mangling

When you write `self.__balance` inside a class, Python **secretly renames it at compile time** to `_ClassName__balance`. This is called **name mangling**, and it's Python's way of making private attributes harder to accidentally overwrite.

So `self.__balance = balance` inside `Account.__init__` actually stores the value under the key `"_Account__balance"` in the object's `__dict__`.

---

## Where does `2` go?

When you write:
```python
ac_new_without_getter.balance = 2
```

Python sees a plain attribute name `balance` — **no double underscore, no mangling**. It simply creates a brand new key `"balance"` in the object's `__dict__`. It has absolutely no connection to `__balance`.

You can prove this:
```python
print(ac_new_without_getter.__dict__)
# {'_Account__balance': 100, 'balance': 2}
```

Two completely separate entries. `2` lives in `"balance"`, `100` lives in `"_Account__balance"`. The private one is never touched.

## The bug in your `balance_set` method

Your setter has two problems:

```python
def balance_set(set, value):      # ← 'set' should be 'self' (just a name, but confusing)
    self.__balance = value         # ← 'self' is not defined here! This will crash.
```

The first parameter of any method is the instance reference — conventionally named `self`. You named it `set`, so `self` doesn't exist inside the method body. It should be:

```python
def balance_set(self, value):
    print("changing Balance")
    self.__balance = value         # this correctly writes to _Account__balance
```

## Quick reference: name mangling rules

| Written in class body | Stored as | Accessible from outside as |
|---|---|---|
| `self.__balance` | `_Account__balance` | `obj._Account__balance` |
| `self._balance` | `_balance` | `obj._balance` (convention only, not enforced) |
| `self.balance` | `balance` | `obj.balance` |

Double underscore (`__`) = mangled and protected. Single underscore (`_`) = just a convention, no enforcement.

## 🔍 `__name__` Special Variable

### Module diagram:
```
my_module
  └── test_1.py
        └── print(__name__)

main.py
  └── from my_module import test_1
        └── print(__name__)
```

### Execution flow:
```
① python main.py → test_1 → print(__name__)  → def / class code → execute
②                 print(__name__)
```

> **`__name__` is a special variable — it holds its execution file name in it.**


## 📄 `__name__` Rules (Page 4)

### Rule ①:
```
If __name__ is present in the SAME file you are executing:
    __name__  =>  "__main__"
```

### Rule ②:
```
If __name__ is NOT present in the current file you are executing:
    __name__  →  that file's name
```
