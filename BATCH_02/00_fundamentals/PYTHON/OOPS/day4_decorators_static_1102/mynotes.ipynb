## 🎨 Decorators

> **Definition:** A Decorator is a **function** that takes **another function** as an **argument** and **returns a new function**. It is used to **enhance the functionality** of the base function.

### Diagram:
```
          ┌──────────┐
          │ function │
          └────┬─────┘
               │ (passed as argument)
               ▼
        function(  )  ──────────→  new function
```

### How it works — Step by step:

#### Step 1 — Base function:
```python
def add():
    a = 1        # ①  → return should be present here
    return a     # ②
```

#### Step 2 — Decorator function:
```python
def run(func):          # ③
    def wrap():         # ②
        print()         # ①
        func()
        print()
    return wrap   → ①
    # return run  → ②
```

## 📄 Decorators — Full Code Example (Page 2)

```python
def add():
    print("Addition Function")       # definition → memory address: 101

def my_func(func):
    def wrap():
        print("Start")
        func()
        print("End")
    return wrap

add = my_func(add)

add()
```

### Memory / Execution Diagram:

```
┌──────────────────────────────────────┐
│  add  ← 101                          │
│  ┌────────────────────────────┐      │
│  │  my_func  ← 102            │      │
│  │   ┌──────────────────┐    │      │
│  │   │   wrap  ← 103    │    │      │
│  │   └──────────────────┘    │      │
│  └────────────────────────────┘      │
└──────────────────────────────────────┘
          ↑ return
```

### Execution trace when `add()` is called:

```
add  → 101
func → 101

my_func(func)
  def wrap():
    print()      ← "Start"
    func()       → 103
    print()      ← "End"
  return wrap
```

### Key insight:
```
add = my_func(add)
         ↑
       ❌ NOT 101 anymore
       ✅ address of a function called as wrap → 103
```

So now when you call `add()` → it actually runs `wrap()` at address 103.

## 🔷 Static Variables & Methods (Page 3)

### Diagram — Memory concept:
```
┌─────────────────────┐
│  1  │  2  │  3      │  ← 3 slots at address 101
└─────────────────────┘
         101

                    101.1
                    101.2
```

*(Visual of multiple objects stacking / pointing to the same memory block)*

### Static — Definition & Properties:

> **Static = belong to the CLASS, not an object**

```
✦ Shared across all objects
✦ Created once
✦ Not tied to any instance
```


### Class diagram:

```python
class A:
    a = 0        # ← static variable  (no self)
    self.b = 1   # ← instance variable
```

```
obj-1 = A()
obj-2 = A()
```

### Memory diagram:
```
┌──────────────────────────────┐
│  Class A:                    │
│  ┌─────────────┐             │
│  │    a = 0    │ ←── link    │
│  └─────────────┘             │
│                              │
│  ┌─────────────┐             │
│  │  self.b = 1 │ ← obj-1    │
│  └─────────────┘             │
│                              │
│  ┌─────────────┐             │
│  │  self.b = 1 │ ← obj-2    │
│  └─────────────┘             │
└──────────────────────────────┘
```

- `a = 0` is **one shared copy** linked to the class
- `self.b` is **separate for each object**
